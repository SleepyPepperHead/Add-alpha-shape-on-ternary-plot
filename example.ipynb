```python
'''example'''

import ternary_alphashape
import matplotlib.pyplot as plt
import ternary

scale = 100
points = [(2, 96, 2), (10, 60, 30), (20, 40, 40), (40, 20, 40), (80, 15, 5), (40, 40, 20),
          (30, 25, 45), (25, 60, 15), (33, 33, 34), (50, 30, 20), (20, 50, 30)]

fig, ax = plt.subplots(figsize=(5.5, 5))

# ternary scatter plot
tax = ternary.TernaryAxesSubplot(ax=ax, scale=scale)
tax.scatter(points, c='green')
tax.boundary(linewidth=2.0)
tax.ticks(axis='rlb', multiple=20, linewidth=1, offset=0.025)
tax.get_axes().axis('off')

# add alpha shape
ternary_alphashape.add_alphashap(points, ax=ax, fc='gold')

plt.show()
```
